```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split
!pip install mlxtend
import joblib
import sys
sys.modules['sklearn.externals.joblib'] = joblib
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from mlxtend.feature_selection import ExhaustiveFeatureSelector as EFS
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
```

```python
delirium_dataset=pd.read_csv('/content/delirium_mimic_dataset_missing_imputed.csv')
```

```python
delirium_dataset.columns
```

```python
delirium_dataset.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(delirium_dataset.drop(labels=['Delirium'], axis=1), 
                                                    delirium_dataset['Delirium'], test_size=0.2,
                                                    random_state=0)
X_train.shape, X_test.shape
```

Forward Selection - Wrapper Approach

```python
sfs1 = SFS(RandomForestClassifier(n_jobs=-1,n_estimators=5), 
           k_features=10, 
           forward=True, 
           floating=False, 
           verbose=1,
           scoring='roc_auc',
           cv=3)

sfs1 = sfs1.fit(np.array(X_train), y_train)
```

```python
selected_feat1= X_train.columns[list(sfs1.k_feature_idx_)]
selected_feat1
```

Backward Elimination - Wrapper Approach

```python
sfs2 = SFS(RandomForestClassifier(n_jobs=-1,n_estimators=5), 
           k_features=10, 
           forward=False, 
           floating=False, 
           verbose=1,
           scoring='roc_auc',
           cv=3)

sfs2 = sfs1.fit(np.array(X_train.fillna(0)), y_train)
```

```python
selected_feat2= X_train.columns[list(sfs2.k_feature_idx_)]
selected_feat2
```
