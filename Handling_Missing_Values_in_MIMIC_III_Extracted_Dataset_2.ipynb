```python
cat_vars=['ADMISSION_TYPE',
       'ADMISSION_LOCATION', 'DISCHARGE_LOCATION', 'INSURANCE',
       'RELIGION', 'MARITAL_STATUS', 'ETHNICITY']
```

```python
def createImputer(colToImpute):
  data_encoder_cols=[]
  data_featurizer_cols =[]
  if colToImpute in cat_vars:
    label_encoder_cols = [datawig.CategoricalEncoder(colToImpute)]
  else:
    label_encoder_cols = [datawig.NumericalEncoder(colToImpute)]
  for i in cols:
    if i !=colToImpute:
      if i in cat_vars:
        data_encoder_cols.append(datawig.CategoricalEncoder(i))
        data_featurizer_cols.append(datawig.EmbeddingFeaturizer(i))
      else:
        data_encoder_cols.append(datawig.NumericalEncoder(i))
        data_featurizer_cols.append(datawig.NumericalFeaturizer(i))
  imputer = datawig.Imputer(
            data_featurizers = data_featurizer_cols,
            data_encoders = data_encoder_cols,
            label_encoders = label_encoder_cols
)
  return imputer
```

```python
def mergeDFs(df,df_imputed,imputed_col):
  l=imputed_col+'_imputed_proba'
  if l not in list(df_imputed):
    df_imputed2=df_imputed.drop(imputed_col,axis=1)
  else:
    df_imputed2=df_imputed.drop([imputed_col,l],axis=1)
  df=df[df[imputed_col].isnull()==False]
  df_imputed2=df_imputed2.rename(columns = {imputed_col+'_imputed':imputed_col})
  return pd.concat([df,df_imputed2])
```

```python
def imputer(df, col):
  missing = df[df[col].isnull()]
  cols=list(df)
  cols.remove(col)
  #cols.remove('Delirium')
  imputer = datawig.SimpleImputer(
    input_columns=cols,
    output_column=col
    )
  imputer.fit(train_df = df)
  imputed = imputer.predict(missing)
  return mergeDFs(df,imputed,col)
```
